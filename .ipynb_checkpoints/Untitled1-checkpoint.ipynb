```python
import numpy as np
import pandas as pd
import numpy.linalg as la
import datetime

import companies
import quarters

start_year = 1995
start_date = datetime.datetime(year=start_year, month=1, day=1)

end_year = 2019
```

```python
returns = pd.read_csv(f'./data/returns/returns_{start_year}-{end_year}.csv').set_index('Unnamed: 0')
rfr = quarters.get_mean_rfr_between_years(start_year, end_year)/100
gdp = quarters.get_gdp_growth_between_years(start_year, end_year)

mu1 = returns.mean()

all_companies = returns.columns

n_companies = returns.shape[1]

returns['avg_rfr'] = rfr
returns['gdp_growth'] = gdp

mu2 = np.array([np.mean(rfr), np.mean(gdp)])

total_points = returns.shape[1]
n_parameters = total_points-n_companies

a = np.array([5.33/100, 0.9301660004562612/100])
```

```python
returns_array = returns.to_numpy().T.astype(float)
returns_array.shape
```

```python
cov_array = np.cov(returns_array)

sigma11 = cov_array[0:n_companies, 0:n_companies]
sigma21 = cov_array[n_companies:total_points, 0:n_companies]
sigma12 = cov_array[0:n_companies, n_companies:total_points]
sigma22 = cov_array[n_companies:total_points, n_companies:total_points]
```

```python
la.inv(sigma)@mu/(np.ones(n_companies)@la.inv(sigma)@mu)
```

```python
mu = mu1 + sigma12@la.inv(sigma22)@(a-mu2)
mu
```

```python
sigma = sigma11 - sigma12@la.inv(sigma22)@sigma21
sigma
```

```python
w_prime_conditional = la.inv(sigma)@mu/(np.ones(n_companies)@la.inv(sigma)@mu)
w_prime = la.inv(sigma11)@mu1/(np.ones(n_companies)@la.inv(sigma11)@mu1)
```

```python
w_prime
```

```python
import yfinance as yf
from dev_email import send_email
import timeit

start_year_local = 2019
start_date_local = datetime.datetime(year=start_year_local, month=1, day=1)

start = timeit.default_timer()

markowitz = pd.DataFrame(index=yf.Ticker('AAPL').history(start=start_date_local).index)
markowitz_conditional = pd.DataFrame(index=yf.Ticker('AAPL').history(start=start_date_local).index)

for i in range(len(all_companies)):
    
    company = all_companies[i]
    weight_cond = w_prime_conditional[i]
    weight = w_prime[i]
    history = yf.Ticker(company).history(start=start_date_local)['Close']
    
    adjusted = history/history.iloc[0]*weight
    adjusted_conditional = history/history.iloc[0]*weight_cond
    
    markowitz[company] = adjusted
    markowitz_conditional[company] = adjusted_conditional

end = timeit.default_timer()
time = end-start
send_email(f'\"Evaluating portfolio using optimal weights between {start_year} and {end_year}. Took {time/60} minutes\"')
```

```python
import plotly.express as px

total_markowitz_conditional = markowitz_conditional.sum(axis=1)
total_markowitz = markowitz.sum(axis=1)
benchmark = yf.Ticker('QQQ').history(start=start_date_local)['Close']

comparison = pd.DataFrame(index=yf.Ticker('AAPL').history(start=start_date_local).index)
comparison['regular'] = total_markowitz
comparison['conditional'] = total_markowitz_conditional
comparison['benchmark'] = benchmark/benchmark.iloc[0]

#px.line(total_portfolio_value, SPY)
px.line(comparison)
```
